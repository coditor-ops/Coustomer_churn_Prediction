```python
X_Data = data.drop(columns=["Churn", "customerID"])
y_Data = data["Churn"]


print(X_Data)

print(y_Data)
```

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, y_Train, y_Test = train_test_split(X_Data, y_Data, test_size=0.2, random_state=1)
print(X_Train)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

y_train = le.fit_transform(y_Train)
y_test = le.transform(y_Test)

print(y_train)


print(y_test)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

cat_cols = [
    "gender",
    "Partner",
    "Dependents",
    "PhoneService",
    "MultipleLines",
    "InternetService",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies",
    "Contract",
    "PaperlessBilling",
    "PaymentMethod"
]
num_cols = [
    "SeniorCitizen",
    "tenure",
    "MonthlyCharges",
    "TotalCharges"
]

processor = ColumnTransformer(
    transformers=[
        ("OHE", OneHotEncoder(), cat_cols),
        ("SS", StandardScaler(), num_cols)
    ]
)

# Fix TotalCharges blank strings
X_Train['TotalCharges'] = pd.to_numeric(X_Train['TotalCharges'], errors='coerce').fillna(0)
X_Test['TotalCharges'] = pd.to_numeric(X_Test['TotalCharges'], errors='coerce').fillna(0)

X_train_processed = processor.fit_transform(X_Train)
X_test = processor.transform(X_Test)

from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train_processed, y_train)
X_test = processor.transform(X_Test)


print(X_train,X_test)
```

```python
from sklearn.svm import SVC
svm_model = SVC(probability=True)
svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)
```

```python
SVM_Train = svm_model.score(X_train, y_train)
SVM_Test = svm_model.score(X_test, y_test)

print(f"SVM Model Performance:")
print(f"Train Accuracy: {SVM_Train:.4f}")
print(f"Test Accuracy: {SVM_Test:.4f}")
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix

SVM_Accuracy = accuracy_score(y_test, svm_predictions)
print(SVM_Accuracy)

SVM_Report = classification_report(y_test, svm_predictions)
print(SVM_Report)


SVM_ConfusionMatrix = confusion_matrix(y_test, svm_predictions)
print(SVM_ConfusionMatrix)
```
