```python
X_Data = data.drop(columns=["Churn", "customerID"], axis = 1)
y_Data = data["Churn"]


print(X_Data)

print(y_Data)
```

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, y_Train, y_Test = train_test_split(X_Data, y_Data, test_size=0.2, random_state=1)
print(X_Train)
```

```python
print(type(X_Train))
```

```python
print(X_Test)
```

```python
print(y_Train)
```

```python
print(y_Test)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

y_train = le.fit_transform(y_Train)
y_test = le.transform(y_Test)

print(y_train)


print(y_test)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

cat_cols = [
    "gender",
    "Partner",
    "Dependents",
    "PhoneService",
    "MultipleLines",
    "InternetService",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies",
    "Contract",
    "PaperlessBilling",
    "PaymentMethod"
]
num_cols = [
    "SeniorCitizen",
    "tenure",
    "MonthlyCharges",
    "TotalCharges"
]

processor = ColumnTransformer(
    transformers=[
        ("OHE", OneHotEncoder(), cat_cols),
        ("SS", StandardScaler(), num_cols)
    ]
)

X_train = processor.fit_transform(X_Train)
X_test = processor.transform(X_Test)


print(X_train,X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

logistic_model = LogisticRegression()
svm_model = SVC()
rfr_model = RandomForestClassifier(max_depth=10)

logistic_model.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
rfr_model.fit(X_train, y_train)

logistic_predictions = logistic_model.predict(X_test)
svm_predictions = svm_model.predict(X_test)
rfr_predictions = rfr_model.predict(X_test)


print("Logistic Regression Predictions:")
print(logistic_predictions)

print("\nSVM Predictions:")
print(svm_predictions)

print("\nRandom Forest Classifier Predictions:")
print(rfr_predictions)
```

```python
Logistic_Train = logistic_model.score(X_train, y_train)
Logistic_Test = logistic_model.score(X_test, y_test)

print(f"Logistic Model Performance:")
print(f"Train Accuracy: {Logistic_Train:.4f}")
print(f"Test Accuracy: {Logistic_Test:.4f}")
```

```python
SVM_Train = svm_model.score(X_train, y_train)
SVM_Test = svm_model.score(X_test, y_test)

print(f"SVM Model Performance:")
print(f"Train Accuracy: {SVM_Train:.4f}")
print(f"Test Accuracy: {SVM_Test:.4f}")
```

```python
RFR_Train = rfr_model.score(X_train, y_train)
RFR_Test = rfr_model.score(X_test, y_test)

print(f"Random Forest Classification Performance:")
print(f"Train Accuracy: {RFR_Train:.4f}")
print(f"Test Accuracy: {RFR_Test:.4f}")
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix

SVM_Accuracy = accuracy_score(y_test, svm_predictions)
print(SVM_Accuracy)

SVM_Report = classification_report(y_test, svm_predictions)
print(SVM_Report)


SVM_ConfusionMatrix = confusion_matrix(y_test, svm_predictions)
print(SVM_ConfusionMatrix)
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix

rfc_Accuracy = accuracy_score(y_test, rfr_predictions)
print(rfc_Accuracy)

rfr_Report = classification_report(y_test, rfr_predictions)
print(rfr_Report)


rfr_ConfusionMatrix = confusion_matrix(y_test, rfr_predictions)
print(rfr_ConfusionMatrix)
```

```python
from sklearn.model_selection import cross_val_score
svm_cv = cross_val_score(svm_model, X_train, y_train)
print(svm_cv.mean())
```

```python
from sklearn.model_selection import cross_val_score
rfr_cv = cross_val_score(rfr_model, X_train, y_train)
print(rfr_cv.mean())
```
